```python
import plotly.express as px
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv("/content/DailyDelhiClimateTest.csv")
df.head()
df.dropna(inplace=True)
df.plot(y ="humidity" )

#Seasonal_decompose
result = seasonal_decompose(df['humidity'], model = 'additive',period = 12)
result.seasonal.plot()
result.trend.plot()
result.plot()


print("form graph, there is no trend and sesonality. Hence data is NON-STATIONARY")
```

**QUESTION 4.**

```python
import plotly.express as px
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller

df = pd.read_csv("/content/DailyDelhiClimateTest.csv")
df.head()
series = df.loc[:,'humidity'].values
fig = px.line(df, x='date', y='humidity')
fig.show()

result4 = adfuller(series, autolag= 'AIC')
print(result4)
print(f'ADF Statistics: {result4[0]}')
print(f'n_lags: {result4[1]}')
print(f'p-value: {result4[1]}')

for key, value in result4[4].items():
  print('Critical Values:')
  print(f'{key}, {value}')
```

**QUESTION 5.**

```python
import plotly.express as px
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import kpss

df = pd.read_csv("/content/DailyDelhiClimateTest.csv")
df.head()

series = df.loc[:,'humidity'].values
fig = px.line(df, x='date', y='humidity')
fig.show()

kpsstest = kpss(series, regression = 'c')
print(kpsstest)
print(f'TEST Statistics: {kpsstest[0]}')
print(f'p-value : {kpsstest[1]}')
print(f'n_lags : {kpsstest[1]}')

for key, value in kpsstest[3].items():
  print('Critical Values:')
  print(f'{key}, {value}')
```
