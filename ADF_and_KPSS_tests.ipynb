**QUESTION 1.**\
**Making a combined function for ADF and KPSS test **

```python
def STATS_adfuller_kpss(series):
  from statsmodels.tsa.stattools import adfuller
  result=adfuller(series,autolag='AIC')
  print(result)
  print(f'ADF Statistics: {result[0]}')
  print(f"n_lags: {result[1]}")
  print(f"p-value: {result[1]}")
  for key,value in result[4].items():
    print("Critical values:")
    print(f'{key},{value}')



  from statsmodels.tsa.stattools import kpss
  kpsstest=kpss(series,regression='c')
  print(kpsstest)
  print(f'TEST Statistics: {kpsstest[0]}')
  print(f"n_lags: {kpsstest[1]}")
  print(f"p-value: {result[1]}")
  for key,value in kpsstest[3].items():
    print("Critical values:")
    print(f'{key},{value}')

  if(result[1]>0.05 and kpsstest[1]<0.05):
    print("Both tests conclude that the series is not stationary -> series is not stationary")
  elif(result[1]<0.05 and kpsstest[1]>0.05):
    print("Both tests conclude that the series is stationary -> series is stationary")
  elif(result[1]>0.05 and kpsstest[1]>0.05):
    print("trend stationary")
  else:
    print("difference stationary")
```
